```python
##Step 1:
##Building the HTTP Connection
import socket
import sys
import requests
import requests_oauthlib
import json
```

```python
##Provide the Twitter API keys obtained after creating an Developer APP in twitter
ACCESS_TOKEN = 'xxxxxxxxxxxxxxxxxxxx'
ACCESS_SECRET = 'xxxxxxxxxxxxxxxxxxxxxx'
CONSUMER_KEY = 'xxxxxxxxxxxxxxxxxxxxxxxx'
CONSUMER_SECRET = 'xxxxxxxxxxxxxxxxxxxxxxxx'
my_auth = requests_oauthlib.OAuth1(CONSUMER_KEY, CONSUMER_SECRET,ACCESS_TOKEN, ACCESS_SECRET)
```

```python
##Filter values to be given here
def get_tweets():
	url = 'https://stream.twitter.com/1.1/statuses/filter.json'
    ##Filter here for Specific Location or Region (New York or San Francisco)
    ##Location value filter for entire United States('-130,-20,100,50') 
    ##Location value filter for New York('-74,-40,-73,41') 
    ##Location value filter for San Francisco('-122.75,-36.8,-121.75,37.8') 
	query_data = [('language', 'en'), ('locations', '-130,-20,100,50'),('track','#')] 
	query_url = url + '?' + '&'.join([str(t[0]) + '=' + str(t[1]) for t in query_data])
	response = requests.get(query_url, auth=my_auth, stream=True)
	print(query_url, response)
	return response
  
```

```python
##Sending the fetched tweets to Spark
def send_tweets_to_spark(http_resp, tcp_connection):
  for line in http_resp.iter_lines():
    try:
      full_tweet = json.loads(line)
      tweet_text = full_tweet['text']
      print("Tweet Text: " + tweet_text)
      print ("------------------------------------------")
      tcp_connection.send(str(tweet_text + '\n').encode())
    except:
      e = sys.exc_info()[0]
      print("This Error: %s" % e)
```

```python
##Setting up Connection
TCP_IP = "localhost"
TCP_PORT = 9099 
conn = None
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.bind((TCP_IP, TCP_PORT))
s.listen(1)
print("Waiting for TCP connection...")
conn, addr = s.accept()
print("Connected... Starting getting tweets.")
resp = get_tweets()
send_tweets_to_spark(resp, conn)
```
