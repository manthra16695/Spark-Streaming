```python
##Visualizing the % Share of Predicted Sentiments in a real time plot
##Importing Matplot Library
%matplotlib notebook
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('percent.csv',header=None) ##File saved from the previous step (After Prediction from ML code)
     
##using the inbuilt index as value for x in plot
    x=data.index
    x1=data.index

    y1=data[data.columns[1]]
    y2 = data[data.columns[2]]
    y3 = data[data.columns[3]]

    plt.cla()

    plt.plot(x, y1, label='Negative Sentiment%')
    plt.plot(x, y2, label='Positive Sentiment%')
    plt.plot(x, y3, label='Neutral Sentiment%')

    plt.legend(loc='upper right')
    plt.tight_layout()

##Checking for the csv file every 1000 Milliseconds
ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()

```
